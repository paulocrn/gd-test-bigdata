```python
# Importar Librerias
import csv
import os
import re
from faker import Faker
import random
from datetime import datetime, timedelta

# Inicializar Faker para creación de data aleatoria
fake = Faker("es_ES")
```

```python
# Setear Cantidad de registros a generar
NUM_CLIENTES = 1000
NUM_PRODUCTOS = 500
NUM_ORDENES = 30000
NUM_DETALLE_ORDENES = 40000
NUM_CORRESPONSALES = 1000
```

```python
# Inicializar listas para almacenar datos
clientes_data = []
productos_data = []
ordenes_data = []
detalle_orden_data = []
corresponsales_data = []


# Generar datos para la tabla 'clientes'
print("Generando datos para la tabla 'clientes'...")
for id in range(1, NUM_CLIENTES + 1):
    nombre = fake.name()
    email = fake.unique.email()
    telefono = fake.phone_number()
    clientes_data.append([id, nombre, email, telefono])



# Generar datos para la tabla 'productos'
print("Generando datos para la tabla 'productos'...")
for id in range(1, NUM_PRODUCTOS + 1):
    nombre = fake.unique.catch_phrase()
    descripcion = fake.sentence()
    precio = round(fake.random.uniform(10, 100), 5)
    productos_data.append([id, nombre, descripcion, precio])



# Generar datos para la tabla 'corresponsales'
print("Generando datos para la tabla 'corresponsales'...")
for id in range(1, NUM_CORRESPONSALES + 1):
    nombre = fake.unique.name()
    direccion = fake.unique.street_name()
    direccion_limpia = re.sub(r'[,]', '', direccion)
    direccion_limpia = re.sub(r'[^a-zA-Z0-9\s]', '', direccion_limpia)
    telefono = fake.phone_number()
    email = fake.unique.email()
    fecha = fake.date_between(start_date="-1y", end_date="today")
    corresponsales_data.append([id, nombre, direccion_limpia, telefono, email, fecha])


cliente_ids = [row[0] for row in clientes_data]
corresponsales_ids = [row[0] for row in corresponsales_data]



# Generar datos para la tabla 'ordenes'
print("Generando datos para la tabla 'ordenes'...")
for id in range(1, NUM_ORDENES + 1):
    fecha = fake.date_between(start_date="-1y", end_date="today")
    #fecha = fake.date_between(start_date="2025-01-01", end_date="2025-01-31")
    id_cliente = random.choice(cliente_ids)
    id_corresponsal = random.choice(corresponsales_ids)
    ordenes_data.append([id, fecha, id_cliente, id_corresponsal])


ordenes_ids = [row[0] for row in ordenes_data]
productos_ids = [row[0] for row in productos_data]

fechas_disponibles = [row[1] for row in ordenes_data]

# Generar datos para la tabla 'detalle_orden'
print("Generando datos para la tabla 'detalle_orden'...")

for id_orden in ordenes_ids:
    fecha = next((row[1] for row in ordenes_data if row[0] == id_orden), None)  # Obtener la fecha correcta
    id_producto = random.choice(productos_ids)
    cantidad = fake.random.randint(1, 20)
    precio = round(fake.random.uniform(10, 100), 5)

    # Asegurar que cada orden tiene al menos un detalle
    detalle_orden_data.append([id_orden, fecha, id_producto, cantidad, precio])

    # Permitir que algunas órdenes tengan más detalles
    if random.random() < 0.3:  # Un 30% de las órdenes tendrán más de un producto
        id_producto_extra = random.choice(productos_ids)
        cantidad_extra = fake.random.randint(1, 10)
        precio_extra = round(fake.random.uniform(10, 100), 5)
        detalle_orden_data.append([id_orden, fecha, id_producto_extra, cantidad_extra, precio_extra])



# Función para escribir datos en archivos CSV
def write_csv(filename, data, headers):
    # Obtener el nombre de la carpeta desde el filename
    folder = os.path.dirname(filename)

    # Crear la carpeta si no existe
    if folder and not os.path.exists(folder):
        os.makedirs(folder)

    
    with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
        csvwriter = csv.writer(csvfile)
        csvwriter.writerow(headers)
        csvwriter.writerows(data)
        

# Escribir datos en archivos CSV
print("Escribiendo datos en archivos CSV...")
write_csv('csv/clientes.csv', clientes_data, ['id', 'nombre', 'email', 'telefono'])
write_csv('csv/productos.csv', productos_data, ['id', 'nombre', 'descripcion', 'precio'])
write_csv('csv/corresponsales.csv', corresponsales_data, ['id', 'nombre', 'direccion', 'telefono', 'email', 'fecha'])
write_csv('csv/ordenes.csv', ordenes_data, ['id', 'fecha', 'id_cliente', 'id_corresponsal'])
write_csv('csv/detalle_orden.csv', detalle_orden_data, ['id_orden', 'fecha', 'id_producto', 'cantidad', 'precio'])

print("Generación de datos y escritura en CSV completada.")
```
